## Project 1 - Predicting Active Users on a Website

Time series analysis involves examining datasets that change over time to identify patterns, trends, and seasonal variations. This type of analysis is used in various fields such as finance, economics, and meteorology to forecast future values based on past data.

- Steps Involved in Time Series Analysis:
  
1- Data Collection: The first step is gathering the relevant data over a specified period. This data must be accurate and timely to ensure meaningful analysis.

2- Visualisation: Once collected, the data is visualised using plots and graphs to understand its underlying patterns, trends, and any seasonal variations.

3- Decomposition: The time series data is decomposed into its constituent components: trend, seasonality, and residual (noise). This helps in better understanding the individual elements affecting the data.

4- Stationarity: For accurate modelling, the data must be stationary, meaning its statistical properties (mean, variance) do not change over time. Techniques such as differencing or transformation are used to achieve stationarity if needed.

5- Model Selection: Various models can be applied to time series data, including ARIMA (AutoRegressive Integrated Moving Average), SARIMA (Seasonal ARIMA), and EMA (Exponential Moving Average). The choice of model depends on the characteristics of the data and the specific requirements of the analysis.

6- Parameter Estimation and Validation: Once a model is selected, its parameters need to be estimated and validated. This involves using techniques like cross-validation to ensure the model performs well on unseen data.

7- Forecasting: Using the validated model, future values of the time series are predicted. This is the core purpose of time series analysis and helps in making informed decisions based on the forecasts.

8- Evaluation and Updating: The accuracy of the forecasts is evaluated using metrics such as Mean Absolute Error (MAE) or Root Mean Square Error (RMSE). Based on the evaluation, the model might need to be updated or refined to improve its forecasting accuracy.

### Statistical Properties of Time Series

Before starting any predictive modeling on time series, it is necessary to check if these statistical properties are constant in the time series:

**Stationarity:** The time series **must** be stationary, meaning its statistical characteristics, such as mean and variance, do not change over time.

**Non-autocorrelation:** The observations of the time series should not be strongly correlated with previous observations.

**Normality:** It is often assumed that the observations of the time series are normally distributed (follow a normal distribution).

### Plot of Rolling Statistics (Moving Statistics)

Rolling statistics, also known as moving statistics or sliding statistics, are a technique used in time series analysis to calculate statistics such as mean, median, sum, variance, and standard deviation over a fixed-size sliding window along the series. The idea is to analyze and understand the changes in local statistics over time, which can help identify trends, seasonality, change points, and other characteristics of the data.

To calculate rolling statistics, you define a window of size 'n' (where n is an integer) and move the window along the time series, calculating the desired statistic for each subset of data within the window. The window advances one observation at a time, and the calculation is repeated until the window reaches the end of the series.

For example, to calculate the moving average of a time series, you select a window of size n and calculate the average of the values within the window. Then, the window is shifted one position forward, and the average is recalculated for the new values in the window. This process is repeated until the window reaches the end of the time series.Plot of Rolling Statistics (Moving Statistics)

* Interpretation of the Graphs:

In the plot, the vertical axis indicates the autocorrelation, and the horizontal axis indicates the lag.

The shaded area indicates where the autocorrelation is significantly different from zero. As can be seen in the image, we have several ACF values (vertical bars) above the shaded area limit. In these cases, the autocorrelation is different from zero, indicating that the series is not random  as expected.

Some vertical bars are within the shaded area limit. This means that the autocorrelation between the series and some of its lags is zero, indicating no correlation.

In simple terms: the shaded area indicates significance. If it is exceeded, it means there is a correlation.

Each bar in the ACF plot represents the level of correlation between the series and its lags in chronological order. The shaded area indicates whether the level of correlation between the series and each lag is significant or not.

The PACF is just a partial visualization of the ACF.

* Analysis:

If the time series is stationary, the ACF/PACF plots will show a "rapid drop in correlation" after a small lag between points. This is not the case in my plots. The slow decline (reduction in the size of the bars) indicates that this series is not stationary.

My series data is not stationary, as a high number of previous observations are correlated with future values.

Confidence intervals are drawn in the shaded area. By default, this is set to a 95% confidence interval, suggesting that correlation values outside this area are likely a correlation and not a statistical fluke.

Here are the key aspects you look for when interpreting an ACF plot:

**1. Temporal Dependence**

ACF measures the correlation between observations of a time series separated by different lags. A significant peak at a specific lag indicates that the values in the time series are correlated with their values at previous lags. This is crucial for identifying whether the series has temporal dependence.

**2. Identification of Seasonality**

If the ACF plot shows regular peaks at specific intervals, this may indicate the presence of seasonality in the series. For example, a significant peak every 12 lags in monthly data may suggest an annual seasonal pattern.

**3. Determination of ARIMA Model Order**

For ARIMA (AutoRegressive Integrated Moving Average) models, analyzing the ACF plot (along with the PACF - Partial Autocorrelation Function) can help determine the order of the AR (autoregressive) and MA (moving average) components. For example, a gradual decline in the ACF suggests AR components, while a sharp cutoff after a specific lag suggests MA components.

**4. Stationarity**

A time series is stationary if its mean and variance remain constant over time. In the ACF plot, a very slow decline or the absence of decline in autocorrelations may indicate that the series is not stationary. Stationarity is often a prerequisite for effective time series modeling.

**5. White Noise**

If all autocorrelations are insignificant (i.e., close to zero for all lags), the series may be white noise, indicating that the values are independently and identically distributed. This means that the series does not have predictable patterns or significant temporal dependence.

### Augmented Dickey-Fuller (ADF) Test

The Augmented Dickey-Fuller (ADF) test is an extended version of the Dickey-Fuller test, used to check for the presence of a unit root in a time series, which means testing the stationarity of the series. A time series is considered stationary when its statistical properties, such as mean and variance, are constant over time. The presence of a unit root indicates that the series is non-stationary, which can affect the analysis and modeling of the series.

The ADF test is particularly useful because it can accommodate a more complex autocorrelation structure compared to the original Dickey-Fuller test. This is done by including lagged terms of the dependent variable in the test equation to control for serial autocorrelation.

These are the hypotheses of the test:

* Null Hypothesis (H0): The series has a unit root (=0), indicating non-stationarity.
* Alternative Hypothesis (H1): The series does not have a unit root (<0 for the one-sided case), indicating stationarity.

Or, more simply:

* Null Hypothesis (H0): The time series is non-stationary.
* Alternative Hypothesis (H1): The time series is stationary.

```python
# Dickey-Fuller Test
print('\nResult Dickey-Fuller Test:\n')

# Test
dftest = adfuller(data_series.active_users, autolag = 'AIC')

# Formatting the output
dfoutput = pd.Series(dftest[0:4], index = ['Test Statistics',
                                        'p-Value',
                                        'Lags Number Considered',
                                        'Observations Number Used'])
# Loop for each item in the test output
for key,value in dftest[4].items():
    dfoutput['Critic Value (%s)'%key] = value

# Print
print(dfoutput)
```


Analysis:

* The p-value is greater than 0.05. **I do not have statistical evidence** to reject the null hypothesis that the series is non-stationary. Therefore, I can conclude that the series is NOT stationary.

## Automating the Stationarity Test

A function that performs the stationarity test and can be applied to any series to test whether it is stationary or not.

**If the series is not stationary, I will need to apply transformations before modeling. In my example, the series is not stationary, and I will transform it into a stationary series later.**

This applies to any time series.

```python
# Function to test stationarity
def zelly_stationarity_test(series, window = 12, title = 'Rolling Average & Dickey-Fuller Test'):
    """
    Function to test the stationarity of the time series.

    Parameters:
    - series: pandas.Series. Time series to be tested.
    - window: int. Window to calculate rolling statistics.
    - title: str. Title for the graphs.
    """
    # Calculating the rolling average
    rolmean = series.rolling(window = window).mean()
    rolstd = series.rolling(window = window).std()

    # Plot Rolling Statistics
    plt.figure(figsize = (14,6))
    plt.plot(series, color = 'blue', label = 'Original')
    plt.plot(rolmean, color = 'red', label = 'Rolling Average')
    plt.plot(rolstd, color ='black', label = 'Rolling Standard Deviation')
    plt.legend(loc = 'best')
    plt.title(title)
    plt.show(block = False)

    # Dickey_Fuller Test
    print('\nResult Dickey-Fuller Test: ')
    dftest = adfuller(series, autolag = 'AIC') #adfuller creates vary models using the AIC metric
    dfoutput = pd.Series(dftest[0:4], index =['Test Statistic',
                                              'p-Value',
                                              'Lags Number Considered',
                                              'Observations Number Used'])
    for key,value in dftest[4].items():
        dfoutput['Critic Value (%s)' % key] = value

    print(dfoutput)

    # Conclusion based in the p-value
    if dfoutput['p-Value'] > 0.05:
        print('\nConclusion:\nThe p-value is greater than 0.05, therefore, we do not have evidence to reject the null hypothesis.\nThis series is probably not stationary.')
    else:
        print('\nConclusion:\nThe p-value is less than 0.05, therefore, we have evidence to reject the null hypothesis.\nThis series is probably stationary.')
              
```

```python
# Executing the function
zelly_stationarity_test(data_series)
```
